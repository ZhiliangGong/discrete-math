```python
def perm(arr):
    if len(arr) == 1:
        return arr
    ans = []
    for i in range(len(arr)):
        if i > 0 and arr[i] == arr[i-1]:
            continue
        sliced = arr[:i] + arr[i+1:]
        res = perm(sliced)
        for r in res:
            ans.append(arr[i] + r)
    return ans
```

```python
arr = ['a', 'a', 'b', 'c', 'f', 'g', 'g']
arr = sorted(arr)
```

```python
perm(arr)
```
